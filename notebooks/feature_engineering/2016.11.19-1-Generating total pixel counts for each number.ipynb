# Generating total pixel counts for each number

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.svm as skl_svm
import sklearn.cross_validation as skl_cv
import seaborn as sns
sns.set_context(context = 'notebook', font_scale=1.5)
import os
import sys
import cv2

import data_processor as mdp
import ellipse_functions as elf
import general_visualizations

data_processor = mdp.MNIST_data_processor()
ellipse_functions = elf.ellipse_functions()
```

```python
X, y = data_processor.load_subset_data()
```

```python
total_pixel_count = X.apply(lambda number_row: number_row.sum(), axis=1)
total_pixel_count.name = 'total_pixel_count'
data = pd.concat([total_pixel_count, y], axis=1)
```

### Comparing 1s and 7s

```python
general_visualizations.compare_distribution_by_number(data = data, 
                                                      column_name='total_pixel_count', 
                                                      labels = [1,7])
```

This looks like it could discriminate between 1s and 7s pretty well

```python
general_visualizations.compare_distribution_by_number(data = data, 
                                                      column_name='total_pixel_count', 
                                                      labels = [2,3,4,5])
```

* 3s are slightly bigger than 4s and 5s
